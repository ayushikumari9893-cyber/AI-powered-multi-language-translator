```python
from deep_translator import GoogleTranslator

translator = GoogleTranslator(source='auto', target='es') # Translate to Spanish
text = "Hello, how are you?"
translated = translator.translate(text)
print(translated)
```

```python
translator_german = GoogleTranslator(source='auto', target='de') # Translate to German
translated_german = translator_german.translate(text)
print(translated_german)
```
